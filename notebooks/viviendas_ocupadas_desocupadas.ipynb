```python
# abrir xlsx
import pandas as pd

ruta = '../data/processed/anexo-proyecciones-viviendas-dptal.xlsx'
df = pd.read_excel(ruta)

# wb.sheetnames
df.head()
```

```python
# dejar solo los valores de area que corresponden a Total
df = df[df['Área'] == 'Total']

df.head()
```

```python
# guardar el dataframe
df.to_csv('../data/processed/proyecciones-total-viviendas-1993-2024.csv')
```

```python
# cambiar formato largo en lugar de ancho
df_long = pd.read_csv('../data/processed/proyecciones-total-viviendas-1993-2024.csv')

df_long.head()
```

```python

# cambiar formato largo
df_long_wide = df_long.melt(id_vars=['Código Departamento', 'Nombre Departamento', 'Área'], var_name= "Año", value_name="Hogares")

# ordenar por departamento y año
df_long_wide = df_long_wide.sort_values(by=["Código Departamento", "Año"])

df_long_wide.head(50)
```

```python
# guardar el dataframe
df_long_wide.to_csv('../data/processed/total-viviendas-1993-2024.csv')
```
