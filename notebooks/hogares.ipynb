# verificar datos de Hogares desde 1993 hasta 2024 por municipio

```python
import pandas as pd

df = pd.read_excel('../data/processed/proyecciones-hogares-1993-2024.xlsx')

df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
# dejar solo los valores de area que corresponden a Total
df = df[df['Área'] == 'Total']

df.head()
```

```python
# guardar el dataframe
df.to_csv('../data/processed/proyecciones-total-hogares-1993-2024.csv')
```

```python
# cambiar formato largo en lugar de ancho
df_long = pd.read_csv('../data/processed/proyecciones-total-hogares-1993-2024.csv')

df_long.head()
```

```python

# cambiar formato largo
df_long_wide = df_long.melt(id_vars=['Código Departamento', 'Nombre Departamento', 'Área'], var_name= "Año", value_name="Hogares")

# ordenar por departamento y año
df_long_wide = df_long_wide.sort_values(by=["Código Departamento", "Año"])

df_long_wide.head(50)
```

```python
# guardar el dataframe
df_long_wide.to_csv('../data/processed/total-hogares-1993-2024.csv')
```
