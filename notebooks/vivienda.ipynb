# Datos para mi hogar

## Contexto y justificación
- Colombia enfrenta un desafío significativo en el acceso a vivienda digna. A pesar de la existencia de programas de subsidios por parte del Ministerio de Vivienda, muchos ciudadanos desconocen las oportunidades disponibles y cómo acceder a ellas. Esta aplicación busca cerrar esa brecha de información y facilitar el acceso a subsidios.
## Objetivo General
- Desarrollar una aplicación que centralice la información sobre los subsidios de vivienda del gobierno, proporcionando a los usuarios herramientas y asistencia para facilitar el proceso de aplicación.
## Objetivos Específicos
•	Ofrecer información clara y accesible sobre los programas de subsidios de vivienda disponibles.
•	Proporcionar rutas guiadas para la aplicación a los subsidios, incluyendo requisitos y documentación necesaria.
•	Implementar un asistente virtual con inteligencia artificial para orientar a los usuarios en el proceso de aplicación.
## Uso de datos abiertos
- utilizará datos abiertos del Ministerio de Vivienda y otras entidades, garantizando que la información sea actual y verificada.
https://www.datos.gov.co/Vivienda-Ciudad-y-Territorio/Subsidios-De-Vivienda-Asignados/h2yr-zfb2/about_data
